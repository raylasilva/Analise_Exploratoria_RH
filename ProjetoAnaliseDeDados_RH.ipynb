*Análise de dados exploratoria de dados de RH *


*  Entendendo o Problema


Analise exploratoria de uma base de informações de funcionarios de uma empresa ficticia.





**IMPORTANDO AS BIBLIOTECAS**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**CARREGANDO CONJUNTO DE DADOS**

```python
df = pd.read_excel('/Base_RH.xlsx')
```

**EXPLORANDO O CONJUNTO DE DADOS: RESUMO**

```python
df.info
df.count()
df.describe()
```

**BOX PLOT DOS SALÁRIOS DOS FUNCIONARIOS**

```python

plt.figure(figsize=(8, 8))
sns.boxplot(df['Valor Salário'])
plt.title('Box Plot do Salário dos Funcionários')
plt.xlabel('Valor Salário')
plt.show()
```

**MAIOR VALOR DE SALARIO E MENOR VALOR DE SALARIO**

```python
Maior = df['Valor Salário'].max()
Menor = df['Valor Salário'].min()

print('O Maior salário é:', Maior, 'Ja o menor é:', Menor)
```

**QUANTIDADE DE FUNCIONÁRIOS POR SEXO**

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Sexo')
plt.xlabel('Sexo')
plt.ylabel('Contagem')
plt.title('Contagem de Funcionários por Sexo')
plt.xticks(rotation=90)
plt.show()
```

**HISTOGRAMA FUNCIONARIOS POR ESTADO CIVIL**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Estado Civil'], bins=20, kde=True)
plt.title('Distribuição do Estado Civil dos Funcionários')
plt.xlabel('Estado')
plt.ylabel('Frequência')
plt.show()
```

**VARIAÇÃO DE SALARIOS AO LONGO DO TEMPO**

```python
df['Data Salário'] = pd.to_datetime(df['Data Salário'])
df.set_index('Data Salário', inplace=True)
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Valor Salário'], marker='o', linestyle='-')
plt.title('Salários ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Valor de Salarios')
plt.grid(True)
plt.show()
```

**MATRIZ DE CORRELAÇÃO**

```python
correlation_matrix = df[['Idade', 'Valor Salário']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação')
plt.show()
```
